```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from xgboost import XGBClassifier

from sklearn.metrics import classification_report, confusion_matrix, roc_auc_score, roc_curve
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score
```

```python
df=pd.read_csv('/content/test_task.csv')
```

```python
df.sample()
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
df['rep_loan_date'] = pd.to_datetime(df['rep_loan_date'])
df['first_loan'] = pd.to_datetime(df['first_loan'])
df['first_overdue_date'] = pd.to_datetime(df['first_overdue_date'], errors='coerce')
```

```python
# Drop columns with too many missing values or less relevance
df.drop(columns=['score_2', 'first_overdue_date', 'dpd_5_cnt', 'dpd_15_cnt', 'dpd_30_cnt'], inplace=True)
```

```python
for col in ['score_1', 'past_billings_cnt', 'close_loans_cnt']:
    df[col] = df[col].fillna(df[col].median())
```

```python
df['gender'] = df['gender'].astype(int)
```

```python
df['federal_district_nm'] = df['federal_district_nm'].astype('category').cat.codes
```

```python
df['loan_relationship_days'] = (df['rep_loan_date'] - df['first_loan']).dt.days
```

```python
payment_type_cols = [f'payment_type_{i}' for i in range(6)]
df['total_payment_type_freq'] = df[payment_type_cols].sum(axis=1)
```

```python
df.drop(columns=['loanKey', 'TraderKey', 'rep_loan_date', 'first_loan'], inplace=True)
```

```python
# Correlation matrix
plt.figure(figsize=(12, 10))
sns.heatmap(df.corr(), annot=True, fmt=".2f", cmap="coolwarm")
plt.title("Correlation Heatmap")
plt.show()
```

```python
# Numeric distributions
numeric_cols = ['score_1', 'past_billings_cnt', 'close_loans_cnt', 'age', 'loan_relationship_days']
df[numeric_cols].hist(bins=30, figsize=(14, 8))
plt.suptitle("Feature Distributions")
plt.tight_layout()
plt.show()
```

```python
# Default rate by gender
sns.barplot(x='gender', y='bad_flag', data=df)
plt.xticks([0, 1], ['Female', 'Male'])
plt.title("Default Rate by Gender")
plt.show()
```

```python
# Default rate by region
plt.figure(figsize=(12, 4))
sns.barplot(x='federal_district_nm', y='bad_flag', data=df)
plt.title("Default Rate by Region Code")
plt.show()
```

```python
X = df.drop(columns=['bad_flag'])
y = df['bad_flag']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, stratify=y, test_size=0.2, random_state=42)
```

```python
from imblearn.over_sampling import SMOTE

smote = SMOTE(random_state=42)
X_train_sm, y_train_sm = smote.fit_resample(X_train, y_train)
```

```python
# Standardize for logistic regression
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

```python
# âœ… Step 5: Train Logistic Regression
model = LogisticRegression(class_weight='balanced', random_state=42)
model.fit(X_train_sm, y_train_sm)
```

```python
y_pred = model.predict(X_test_scaled)
y_prob = model.predict_proba(X_test_scaled)[:, 1]
```

```python
print("=== Classification Report ===")
print(classification_report(y_test, y_pred))
```

```python
print("=== Confusion Matrix ===")
print(confusion_matrix(y_test, y_pred))
```

```python
print("=== ROC AUC ===")
print("AUC Score:", roc_auc_score(y_test, y_prob))
```

```python
# âœ… Step 7: ROC Curve
fpr, tpr, _ = roc_curve(y_test, y_prob)
plt.figure(figsize=(8, 5))
plt.plot(fpr, tpr, label=f"AUC = {roc_auc_score(y_test, y_prob):.2f}")
plt.plot([0, 1], [0, 1], 'k--')
plt.title("ROC Curve")
plt.xlabel("False Positive Rate")
plt.ylabel("True Positive Rate")
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
from sklearn.metrics import precision_recall_curve
```

```python
# âœ… Step 8: Precision-Recall Curve
precision, recall, thresholds = precision_recall_curve(y_test, y_prob)
plt.figure(figsize=(8, 5))
plt.plot(thresholds, precision[:-1], label="Precision")
plt.plot(thresholds, recall[:-1], label="Recall")
plt.title("Precision and Recall vs Threshold")
plt.xlabel("Threshold")
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
threshold_df = pd.DataFrame({'prob': y_prob, 'true': y_test.values})
thresholds_range = np.arange(0.1, 1.01, 0.01)
default_rates = []

for threshold in thresholds_range:
    approved = threshold_df[threshold_df['prob'] >= threshold]
    default_rate = approved['true'].mean() if len(approved) > 0 else np.nan
    default_rates.append(default_rate)

plt.figure(figsize=(10, 6))
plt.plot(thresholds_range, default_rates, marker='o')
plt.title("Expected Default Rate vs Approval Threshold")
plt.xlabel("Approval Threshold")
plt.ylabel("Observed Default Rate")
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
# âœ… Step 1: Import XGBoost
from xgboost import XGBClassifier

# âœ… Step 2: Apply SMOTE to training data
from imblearn.over_sampling import SMOTE
smote = SMOTE(random_state=42)
X_train_sm, y_train_sm = smote.fit_resample(X_train, y_train)

# âœ… Step 3: Train XGBoost model
xgb_model = XGBClassifier(use_label_encoder=False, eval_metric='logloss', random_state=42)
xgb_model.fit(X_train_sm, y_train_sm)

# âœ… Step 4: Predict on test set
y_pred_xgb = xgb_model.predict(X_test)
y_prob_xgb = xgb_model.predict_proba(X_test)[:, 1]
```

```python
# âœ… Step 5: Evaluation
print("=== XGBoost Classification Report ===")
print(classification_report(y_test, y_pred_xgb))
```

```python
print("=== Confusion Matrix ===")
print(confusion_matrix(y_test, y_pred_xgb))
```

```python
print("=== ROC AUC Score ===")
print(roc_auc_score(y_test, y_prob_xgb))
```

```python
fpr, tpr, _ = roc_curve(y_test, y_prob_xgb)
plt.figure(figsize=(8, 5))
plt.plot(fpr, tpr, label=f"XGBoost (AUC = {roc_auc_score(y_test, y_prob_xgb):.2f})")
plt.plot([0, 1], [0, 1], 'k--')
plt.xlabel("False Positive Rate")
plt.ylabel("True Positive Rate")
plt.title("XGBoost ROC Curve")
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
# Feature Importance Plot
importance_df = pd.DataFrame({
    'Feature': X.columns,
    'Importance': xgb_model.feature_importances_
}).sort_values(by='Importance', ascending=False)

plt.figure(figsize=(10, 6))
sns.barplot(data=importance_df, x='Importance', y='Feature', palette='viridis')
plt.title("XGBoost Feature Importances")
plt.tight_layout()
plt.show()
```
